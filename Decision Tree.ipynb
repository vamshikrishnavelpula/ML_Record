## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
df["Embarked"].fillna("S", inplace=True)
print(df["Embarked"].isnull().sum())
```

```python
#Impute the missing values in Fare column with median value
df["Fare"].fillna(df["Fare"].median(), inplace=True)

print(df["Fare"].isnull().sum())
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
df["Age"].fillna(df["Age"].median(), inplace=True)
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
mean_age = df["Age"].mean()
std_age = df["Age"].std()

# Count missing values
num_missing = df["Age"].isnull().sum()

# Generate random ages from the normal distribution
random_ages = np.random.normal(loc=mean_age, scale=std_age, size=num_missing)

# Fill missing Age values with these random values
df.loc[df["Age"].isnull(), "Age"] = random_ages

# Verify if missing values are replaced
print(df["Age"].isnull().sum())
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df[["SibSp", "Parch", "FamilySize"]].head()
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
df["IsAlone"] = df["FamilySize"] == 1
df[["FamilySize", "IsAlone"]].head()
```

## Create a new attribute Has_cabin using the attribue Cabin. if Cabin == NaN the Has_cabin = False else Has_cabin = True

```python
df["Has_cabin"] = df["Cabin"].notna()

# Print first few rows to check
print(df[["Cabin", "Has_cabin"]].head())
```

```python
#Drop the columns which may not be useful in prediction of survival
columns_to_drop = ["PassengerId", "Name", "Ticket", "Cabin"]

# Drop columns
df_cleaned = df.drop(columns=columns_to_drop)

# Print remaining columns
print("Remaining columns:", df_cleaned.columns.tolist())
```

## Model Building

```python
#Prepare X and y
df_model = df.drop(['Name', 'Ticket', 'Cabin'], axis=1)

# Split into features and target
X = df_model.drop('Survived', axis=1)
y = df_model['Survived']
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = df['Embarked'].fillna('Missing')
df['Embarked'] = le.fit_transform(df['Embarked'])
df[['Sex', 'Embarked']].head()
```

```python
# concat df_categorical with original df
categorical_cols = ['Sex', 'Embarked']
df['Embarked'] = df['Embarked'].fillna('Missing')
df_categorical = pd.DataFrame()

le = LabelEncoder()
for col in categorical_cols:
    df_categorical[col] = le.fit_transform(df[col])

# Concatenate with original DataFrame
df_combined = pd.concat([df, df_categorical.add_suffix('_encoded')], axis=1)

# View the result
df_combined.head()
```

```python
#Do a pairplot of all features
sns.pairplot(df)
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y
)
print("X_train shape:", X_train.shape)
print("X_validation shape:", X_validation.shape)
print("y_train shape:", y_train.shape)
print("y_validation shape:", y_validation.shape)
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create the model
model = DecisionTreeClassifier(random_state=42)

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on validation set
y_pred = model.predict(X_validation)

# Evaluate the model
print("Accuracy on validation set:", accuracy_score(y_validation, y_pred))
print("\nClassification Report:\n", classification_report(y_validation, y_pred))
```

```python
#Print the tree
#Importing required packages for visualization
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 10))

# Plot the decision tree
plot_tree(model, 
          feature_names=X_train.columns, 
          class_names=["Not Survived", "Survived"], 
          filled=True, 
          rounded=True, 
          fontsize=10)

# Show the plot
plt.show()
```

```python
# plotting tree with max_depth=3
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Plotting the tree with max_depth=3
plt.figure(figsize=(15, 8))
plot_tree(dt_model, 
          filled=True, 
          feature_names=X_train.columns, 
          class_names=["Not Survived", "Survived"], 
          max_depth=3, 
          rounded=True, 
          fontsize=10)
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

# Predict on validation set
y_pred = dt_model.predict(X_validation)

# Confusion Matrix
cm = confusion_matrix(y_validation, y_pred)
print("Confusion Matrix:\n", cm)

# Accuracy
acc = accuracy_score(y_validation, y_pred)
print("\nAccuracy:", acc)

# Precision
precision = precision_score(y_validation,y_pred)
print("Precision:", precision)

# Recall
recall = recall_score(y_validation, y_pred)
print("Recall:", recall)

# F1 Score
f1 = f1_score(y_validation, y_pred)
print("F1 Score:", f1)

# Optional: Full classification report
print("\nClassification Report:\n", classification_report(y_validation, y_pred, target_names=["Not Survived", "Survived"]))
```
