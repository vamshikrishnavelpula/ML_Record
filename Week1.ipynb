### 27. Subset rows where an external array, bar, is positive

```python
import numpy as np
bar = np.array([1, -2, 3, 0, 5, -1, 2, -3, 4, 6]) 

if len(bar) != len(df):
    raise ValueError("The length of bar must match the number of rows in df")

df_subset = df[bar > 0]
df_subset
```

### 28. Subset rows where foo is TRUE or bar is negative

```python
foo = np.array([True, False, True, False, True, True, False, True, False, True])  
bar = np.array([1, -2, 3, 0, 5, -1, 2, -3, 4, 6]) 


if len(foo) != len(df) or len(bar) != len(df):
    raise ValueError("The length of foo and bar must match the number of rows in df")


df_subset = df[(foo) | (bar < 0)]
df_subset
```

### 29. Subset the rows where foo is not TRUE and bar is not negative

```python
foo = np.array([True, False, True, False, True, True, False, True, False, True])  
bar = np.array([1, -2, 3, 0, 5, -1, 2, -3, 4, 6])  

if len(foo) != len(df) or len(bar) != len(df):
    raise ValueError("The length of foo and bar must match the number of rows in df")

df_subset = df[(~foo) & (bar >= 0)]
df_subset
```
