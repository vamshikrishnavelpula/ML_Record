```python
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # The last column (Species)
```

```python
from sklearn.model_selection import train_test_split
# Assuming X and y are already defined
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix

# Step 1: Create the model
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("confusion matrix :",confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the model

# Define the parameter grid for C (try multiple values)
param_grid = {'C': [0.1, 1, 10, 100, 1000]}

# Set up GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5)

# Fit to the entire dataset
grid_search.fit(X, y)

# Print the best parameters and score
print("Best C value:", grid_search.best_params_['C'])
print("Best cross-validation score:", grid_search.best_score_)
```

```python
C_values = [0.001, 0.01, 0.1, 1, 10, 100]
param_grid = {'C': C_values}

# Set up GridSearchCV to return training scores too
grid = GridSearchCV(model, param_grid, cv=5, return_train_score=True)
grid.fit(X, y)

# Extract train and test scores
train_scores = grid.cv_results_['mean_train_score']
test_scores = grid.cv_results_['mean_test_score']

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(C_values, train_scores, label='Train Score', marker='o')
plt.plot(C_values, test_scores, label='Test Score', marker='s')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Best C value:", grid.best_params_['C'])
print("Best cross-validation score:", grid.best_score_)
```
